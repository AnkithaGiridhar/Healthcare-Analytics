```python
import numpy as np 
import pandas as pd 
```

```python
#data_dict = pd.read_excel('../input/janatahack-healthcare-analytics/Train/Data_Dictionary.xlsx')
first_hc = pd.read_csv('../input/janatahack-healthcare-analytics/Train/First_Health_Camp_Attended.csv')
hc_detail = pd.read_csv('../input/janatahack-healthcare-analytics/Train/Health_Camp_Detail.csv')
patient_profile = pd.read_csv('../input/janatahack-healthcare-analytics/Train/Patient_Profile.csv')
second_hc = pd.read_csv('../input/janatahack-healthcare-analytics/Train/Second_Health_Camp_Attended.csv')
third_hc = pd.read_csv('../input/janatahack-healthcare-analytics/Train/Third_Health_Camp_Attended.csv')
train_1 = pd.read_csv('../input/janatahack-healthcare-analytics/Train/Train.csv')
test = pd.read_csv('../input/janatahack-healthcare-analytics/Test.csv')
```

```python
set(first_hc['Patient_ID']) - set(train_1['Patient_ID'])
set(second_hc['Patient_ID']) - set(train_1['Patient_ID'])
set(third_hc['Patient_ID']) - set(train_1['Patient_ID'])
set(first_hc['Health_Camp_ID']) - set(train_1['Health_Camp_ID'])
set(second_hc['Health_Camp_ID']) - set(train_1['Health_Camp_ID'])
set(third_hc['Health_Camp_ID']) - set(train_1['Health_Camp_ID'])
len(set(first_hc['Health_Camp_ID']) - set(second_hc['Health_Camp_ID']))
len(set(first_hc['Health_Camp_ID']) - set(third_hc['Health_Camp_ID']))
len(set(second_hc['Health_Camp_ID']) - set(third_hc['Health_Camp_ID']))
len(set(second_hc['Health_Camp_ID']) - set(first_hc['Health_Camp_ID']))
len(set(third_hc['Health_Camp_ID']) - set(first_hc['Health_Camp_ID']))
len(set(third_hc['Health_Camp_ID']) - set(second_hc['Health_Camp_ID']))
#second_hc['Health_Camp_ID'].nunique()
patient_profile['City_Type'].unique()
hc_detail['Category2'].unique()
first_hc[first_hc['Unnamed: 4'].isnull() == True]
first_hc = first_hc.drop(['Unnamed: 4'],axis = 1)
patient_profile[patient_profile['Employer_Category'].isnull() == True]
patient_profile['Employer_Category'].unique()
```

```python
first_hc = first_hc.rename({'Health_Score':'Health_Score_1'}, axis = 1)
second_hc = second_hc.rename({'Health Score':'Health_Score_2'}, axis = 1)
patient_profile['Employer_Category'] = patient_profile['Employer_Category'].fillna('Others')
hc_detail['Category1'] = hc_detail['Category1'].replace({'First':1,'Second':2,'Third':3})
```

```python
t1 = set(train_1['Patient_ID'].astype(str) + train_1['Health_Camp_ID'].astype(str))
t2 = set(first_hc['Patient_ID'].astype(str) + first_hc['Health_Camp_ID'].astype(str))
t3 = set(second_hc['Patient_ID'].astype(str) + second_hc['Health_Camp_ID'].astype(str))
t4 = set(third_hc['Patient_ID'].astype(str) + third_hc['Health_Camp_ID'].astype(str))
print(t2-t1,t3-t1,t4-t1)
t1 = set(test['Patient_ID'].astype(str) + test['Health_Camp_ID'].astype(str))
t2 = set(first_hc['Patient_ID'].astype(str) + first_hc['Health_Camp_ID'].astype(str))
t3 = set(second_hc['Patient_ID'].astype(str) + second_hc['Health_Camp_ID'].astype(str))
t4 = set(third_hc['Patient_ID'].astype(str) + third_hc['Health_Camp_ID'].astype(str))
print(len(t2-t1),len(t3-t1),len(t4-t1))
p1 = set(test['Patient_ID'].unique())
p2 = set(train_1['Patient_ID'].unique())
print(len(p1-p2),len(p2-p1),len(p1),len(p2))
```

```python
print(train_1.shape)
train_2 = train_1.merge(first_hc, on = ['Patient_ID','Health_Camp_ID'],how = 'left')
print(train_2.shape)
train_2 = train_2.merge(second_hc, on = ['Patient_ID','Health_Camp_ID'],how = 'left')
print(train_2.shape)
train_2 = train_2.merge(third_hc, on = ['Patient_ID','Health_Camp_ID'],how = 'left')
print(train_2.shape)
train_2 = train_2.merge(hc_detail, on = ['Health_Camp_ID'],how = 'left')
print(train_2.shape)
train_2 = train_2.merge(patient_profile, on = ['Patient_ID'],how = 'left')
print(train_2.shape)
```

```python
for col in train_2.columns:
    if train_2[col].dtypes != 'object':
        print(col)
        train_2[col].fillna(0,inplace = True)
    else:
        print(len(train_2[train_2[col].isnull()==True]))
#train_2['Donation'].fillna(0,inplace = True)
#train_2['Health_Score_1'].value_counts().index.values.min()
#train_2['Health_Score_1'].min()#fillna(0,inplace = True)
#train_2['Health_Score_1'].values.min()
train_2
```

```python
print(train_2.columns.to_list())
train_2['Outcome'] = np.where((train_2['Health_Score_1']>0) | (train_2['Health_Score_2']>0) | (train_2['Number_of_stall_visited']>0), 1, 0)
train_2['Outcome'].sum()
train_2['type'] = 'train'
train_2.drop(['Health_Score_1', 'Health_Score_2','Number_of_stall_visited', 'Last_Stall_Visited_Number'],axis = 1)
```

```python
print(test.shape)
test_2 = test.merge(first_hc, on = ['Patient_ID','Health_Camp_ID'],how = 'left')
print(test_2.shape)
test_2 = test_2.merge(second_hc, on = ['Patient_ID','Health_Camp_ID'],how = 'left')
print(test_2.shape)
test_2 = test_2.merge(third_hc, on = ['Patient_ID','Health_Camp_ID'],how = 'left')
print(test_2.shape)
test_2 = test_2.merge(hc_detail, on = ['Health_Camp_ID'],how = 'left')
print(test_2.shape)
test_2 = test_2.merge(patient_profile, on = ['Patient_ID'],how = 'left')
print(test_2.shape)
test_2['type'] = 'test'
test_2['Outcome'] = np.nan
```

```python
master = pd.concat([train_2,test_2])
```

```python
for col in master.columns:
    if master[col].dtypes != 'object':
        #print(col)
        master[col].fillna(0,inplace = True)
    else:
        print(col, len(master[master[col].isnull()==True]))
```

```python
master['Registration_Date'].fillna('None',inplace = True)
master['City_Type'].fillna('None',inplace = True)
```
